Use numerical methods to solve for shape and scale parameters of gamma distribution

```python
#import for digamma and trigamma function
import scipy.special as sp
```

```python
# Must used numerical methods
# Newtons method

#Create functions
def f(k, phi, ln_Xbar):
  k = float(k)
  function = sp.digamma(k) - np.log(k) - phi + ln_Xbar
  return function

def df(k):
  k = float(k)
  function = sp.polygamma(1, k) - 1/k
  return function

def newtons_method(kstart, eps, phi, ln_Xbar):
  epsilon = float(eps)
  kn = float(kstart)
  kn1 = 0
  iterations = 0
  while abs(kn-kn1) > epsilon:
    if iterations != 0:
      kn = kn1
    kn1 = kn - f(kn, phi, ln_Xbar)/df(kn)
    iterations += 1
  print("Iterations: ", iterations)
  return kn1
```

```python
# AAPL solve for k = shape 
k_aapl_1 = newtons_method(1, 0.00001, A_phi_1, A_ln_Xbar_1)
print("AAPL hour 1 Shape parameter k = ", np.round(k_aapl_1, 4))
print("\n")

k_aapl_2 = newtons_method(1, 0.00001, A_phi_2, A_ln_Xbar_2)
print("AAPL hour 2 Shape parameter k = ", np.round(k_aapl_2, 4))
print("\n")

k_aapl_3 = newtons_method(1, 0.00001, A_phi_3, A_ln_Xbar_3)
print("AAPL hour 3 Shape parameter k = ", np.round(k_aapl_3, 4))
print("\n")

k_aapl_4 = newtons_method(1, 0.00001, A_phi_4, A_ln_Xbar_4)
print("AAPL hour 4 Shape parameter k = ", np.round(k_aapl_4, 4))
print("\n")

k_aapl_5 = newtons_method(1, 0.00001, A_phi_5, A_ln_Xbar_5)
print("AAPL hour 5 Shape parameter k = ", np.round(k_aapl_5, 4))
print("\n")

k_aapl_6 = newtons_method(1, 0.00001, A_phi_6, A_ln_Xbar_6)
print("AAPL hour 6 Shape parameter k = ", np.round(k_aapl_6, 4))
print("\n")

k_aapl_7 = newtons_method(1, 0.00001, A_phi_7, A_ln_Xbar_7)
print("AAPL hour 7 Shape parameter k = ", np.round(k_aapl_7, 4))
```

Check other distribution - Log-Normal

```python
# MLE for log- normal distribution
# mu = 1/n * sum(ln(xi))
# sigma**2 = 1/n sum((ln(xi)-mu)**2)

# AAPL
mu_aapl = A_phi
aapl_sigma_sq_list = [(np.log(x)-mu_aapl)**2 for x in aapl_list]

aapl_sigma_sq = sum(aapl_sigma_sq_list)/len(aapl_sigma_sq_list)
aapl_sigma = np.sqrt(aapl_sigma_sq)

# PFE 
mu_pfe = p_phi
pfe_sigma_sq_list = [(np.log(x)-mu_pfe)**2 for x in pfe_list]

pfe_sigma_sq = sum(pfe_sigma_sq_list)/len(pfe_sigma_sq_list)
pfe_sigma = np.sqrt(pfe_sigma_sq)

# JPM 
mu_jpm = j_phi
jpm_sigma_sq_list = [(np.log(x)-mu_jpm)**2 for x in jpm_list]

jpm_sigma_sq = sum(jpm_sigma_sq_list)/len(jpm_sigma_sq_list)
jpm_sigma = np.sqrt(jpm_sigma_sq)
```
